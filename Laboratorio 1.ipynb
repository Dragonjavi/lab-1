```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Problema 1

```python
def filtro_color(imagen, color):
    img = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

    #Dimensiones 
    alto = img.shape[0]
    ancho = img.shape[1]
    canales = img.shape[2]
    # colores vector
    colores = np.zeros_like(img)
    # pixeles
    for i in range(0,alto):
        for j in range(0, ancho):
            pizel = img[i,j]
        
            blue = pizel[0]
            green = pizel[1]
            red = pizel[2]
            # filtrado por colores
            if color == 1:
                colores[i,j] = (blue,0,0)
            elif color == 2:
                colores[i,j] = (0,green,0)
            elif color == 3:
                colores[i,j] = (0,0,red)
            elif color == 10:
                colores[i,j] = (0,green,red)
            elif color == 20:
                colores[i,j] = (blue,green,0)
            elif color == 30:
                colores[i,j] = (blue,0,red)
    return colores


```

```python
# muestreo de colores
img = cv2.imread("Torre.jpg")

color_deseado = 3

# Obtenemos la imagen filtrada según el color seleccionado
imagen_filtrada = filtro_color(img, color_deseado)

# Mostramos la imagen original y la imagen filtrada
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imagen_filtrada)
plt.title('Imagen Filtrada')
plt.axis('off')

plt.show()
```

```python
def construir_imagen_color(img_r, img_g, img_b):
    # Verificar que las imágenes tengan las mismas dimensiones
    if img_r.shape != img_g.shape or img_r.shape != img_b.shape:
        raise ValueError("Las imágenes deben tener las mismas dimensiones.")

    # Crear una matriz 3D para la imagen a color
    img_color = np.zeros((img_r.shape[0], img_g.shape[1], 3), dtype=np.uint8)

    # Asignar cada canal de color a la imagen a color
    img_color[:, :, 0] = img_b  # Canal azul
    img_color[:, :, 1] = img_g  # Canal verde
    img_color[:, :, 2] = img_r  # Canal rojo

    return img_color

# Ejemplo de uso:
img_r = cv2.imread('imagen2_salida_gray_azul.jpg', cv2.IMREAD_GRAYSCALE)
img_g = cv2.imread('imagen2_salida_gray_verde.jpg', cv2.IMREAD_GRAYSCALE)
img_b = cv2.imread('imagen2_salida_gray_rojo.jpg', cv2.IMREAD_GRAYSCALE)

imagen_color = construir_imagen_color(img_r, img_g, img_b)

# Mostrar la imagen a color
plt.imshow(imagen_color)
plt.title('Imagen a color')
plt.axis('off')
plt.show()

```

```python
def construir_imagen_color(img_r, img_g, img_b):
    # Verificar que las imágenes tengan las mismas dimensiones
    if img_r.shape != img_g.shape or img_r.shape != img_b.shape:
        raise ValueError("Las imágenes deben tener las mismas dimensiones.")

    # Crear una matriz 3D para la imagen a color
    img_color = cv2.merge([img_r, img_g, img_b])

    # Asignar cada canal de color a la imagen a color
    img_color[:, :, 0] = img_b  # Canal azul
    img_color[:, :, 1] = img_g  # Canal verde
    img_color[:, :, 2] = img_r  # Canal rojo

    return img_color

# Ejemplo de uso:
img_r = cv2.imread('imagen2_salida_gray_rojo.jpg', cv2.IMREAD_GRAYSCALE)
img_g = cv2.imread('imagen2_salida_gray_verde.jpg', cv2.IMREAD_GRAYSCALE)
img_b = cv2.imread('imagen2_salida_gray_azul.jpg', cv2.IMREAD_GRAYSCALE)

imagen_color = construir_imagen_color(img_r, img_g, img_b)

# Mostrar la imagen a color
plt.imshow(imagen_color)
plt.title('Imagen a color')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def crear_escala_grises_en_tres_dimensiones(imagen_gris):
    # Crear tres copias de la imagen en escala de grises
    img_r = imagen_gris.copy()
    img_g = imagen_gris.copy()
    img_b = imagen_gris.copy()

    # Crear una imagen en escala de grises en tres dimensiones
    img_gris_3d = np.zeros((imagen_gris.shape[0], imagen_gris.shape[1], 3), dtype=np.uint8)

    # Asignar cada copia de la imagen en escala de grises a cada canal de color
    img_gris_3d[:, :, 0] = img_b  # Canal azul
    img_gris_3d[:, :, 1] = img_g  # Canal verde
    img_gris_3d[:, :, 2] = img_r  # Canal rojo

    # Devolver las tres imágenes
    return img_r, img_g, img_b

# Ejemplo de uso:
imagen_gris = cv2.imread('torre.jpg', cv2.IMREAD_GRAYSCALE)

# Crear las tres imágenes en escala de grises en tres dimensiones
img_r, img_g, img_b = crear_escala_grises_en_tres_dimensiones(imagen_gris)

# Mostrar las tres imágenes resultantes utilizando plt
plt.subplot(131)
plt.imshow(img_r, cmap='gray')
plt.title('Canal rojo')
plt.axis('off')

plt.subplot(132)
plt.imshow(img_g, cmap='gray')
plt.title('Canal verde')
plt.axis('off')

plt.subplot(133)
plt.imshow(img_b, cmap='gray')
plt.title('Canal azul')
plt.axis('off')

plt.show()
```
